# Consonance and Dissonance

It is said that Pythagoras' ever first became interested in establishing the relationship between numbers and sounds when he walked in the street and heard a blacksmith hammering. The heavier the metal sheet is, the lower the sound.

Pythagoras went on to make a single-string instrument. (Unlike lyre, which usually has 4, 7 or 10 strings - don't you just always want to start with the simplest model that there is?) He found out that if he vibrates the string at its full length then followed by a vibration at half of its length, then the sounds are particularly pleasant. Moreover, as long as the ratio of the string length are smaller, simple numbers (1:2, 2:3, and 3:4), the combinations make harmonious sound - consonance. With larger ratios, the sounds made are dissonance.

Reflected on the pitches, consonances are those notes kind of "far apart": octaves, perfect fourths and fifths, .etc. dissonances are those notes so close that they are next to each other, or next to the note's octaves.  

```python
from __future__ import print_function
import numpy as np
import scipy
import matplotlib.pyplot as plt

import librosa
import librosa.display
```

```python
consonant_y, consonant_sr = librosa.load('../media/consonant-dissonant-music.mp3', offset=0, duration=15)
dissonant_y, dissonant_sr = librosa.load('../media/consonant-dissonant-music.mp3', offset=16, duration=15)


con_chroma_os = librosa.feature.chroma_cqt(y=consonant_y, sr=consonant_sr, bins_per_octave=12*3)
dis_chroma_os = librosa.feature.chroma_cqt(y=dissonant_y, sr=dissonant_sr, bins_per_octave=12*3)


plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
librosa.display.specshow(con_chroma_os, y_axis='chroma')
plt.colorbar()
plt.ylabel('Consonant')
plt.tight_layout()

plt.subplot(2, 1, 2)
librosa.display.specshow(dis_chroma_os, y_axis='chroma',  x_axis='time')
plt.colorbar()
plt.ylabel('Dissonant')
plt.tight_layout()
```

Chromagrams shows the pitch classes the music falls into against time. In one pitch class, all pitices are a whole number of octaves apart.

We can see how much closer the pitch classes are in the dissonance's chromagram. 

```python
schoenberg_y, schoenberg_sr = librosa.load('../media/schoenberg-11-3-sanchez.mp3')
mozart_y, mozart_sr = librosa.load('../media/mozart-kv279-3-hammer.mp3')

sch_chroma_os = librosa.feature.chroma_cqt(y=schoenberg_y, sr=schoenberg_sr, bins_per_octave=12*3)
moz_chroma_os = librosa.feature.chroma_cqt(y=mozart_y, sr=mozart_sr, bins_per_octave=12*3)


plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
librosa.display.specshow(moz_chroma_os, y_axis='chroma')
plt.colorbar()
plt.ylabel('Mozart')
plt.tight_layout()

plt.subplot(2, 1, 2)
librosa.display.specshow(sch_chroma_os, y_axis='chroma',  x_axis='time')
plt.colorbar()
plt.ylabel('Schoenberg')
plt.tight_layout()
```
