# Harmonic and Percussive

Harmonic sounds are what we hear as melodies and chords. Percussive ones sound like noises and ususally produced from instrument onsets.
Visually, on spectrographs, an ideal harmonic sound is a horizontal line, persisting at a certain frequency over time. While an ideal percussive sound is a vertical line, a trigger of all frequencies at an instant.

Usually all music contains both harmonic and percussive sounds. One way to separate them is Harmonic Percussive Sound Separation (HPSS).

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

import librosa
import librosa.display


y, src = librosa.load('../media/mozart-kv279-3-hammer.mp3', offset=10, duration=10)

D = librosa.stft(y)

D_harmonic, D_percussive = librosa.decompose.hpss(D)

type(D)

rp = np.max(np.abs(D))

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(D_harmonic, ref=rp), y_axis='log', cmap='viridis')
plt.colorbar()
plt.title('Harmonic spectrogram')

plt.subplot(2, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(D_percussive, ref=rp), y_axis='log', x_axis='time', cmap='viridis')
plt.colorbar()
plt.title('Percussive spectrogram')
plt.tight_layout()
```
